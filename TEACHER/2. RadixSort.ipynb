```python
def bubbleSort(arr):
    n = len(arr)
    count = 0
    
    for i in range(n-1):
        swap = False
        for j in range(0, n-i-1):
          count += 1
          if arr[j] > arr[j + 1] :
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swap = True
        if not swap:
          break      
    print("count : ",count)
    
arr = [64, 34, 25, 12, 22, 11, 90]
arr1 = [11, 12, 22, 25, 34, 64, 90]

bubbleSort(arr1)
  
print ("Sorted array is:")
for i in range(len(arr1)):
    print ("% d" % arr1[i]), 
```
