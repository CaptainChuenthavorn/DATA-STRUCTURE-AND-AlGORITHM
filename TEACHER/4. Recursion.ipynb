Recursion

```python
def fiboR(n) :
  if n == 0 or n == 1 :
    return n
  else :
    return fiboR(n-1) + fiboR(n-2)
```

```python
def fiboI(n) :
  if n == 0 or n == 1 :
    return n
  else :
    lo,hi = 0,1
    for i in range (2,n+1):
      new = lo + hi
      lo = hi
      hi = new
    return new
```

```python
for i in range(51) :
  print(i,fiboI(i))
```

```python
for i in range(51) :
  print(i,fiboR(i))
```

```python
def move(n,A,C,B) :
  if n == 1 :
    print(n, 'from', A, 'to', C)
  else :
    move(n-1,A,B,C)
    print(n, 'from', A, 'to', C)
    move(n-1,B,C,A)


move(3,'A','C','B')
```
